```python
list_next_two_glasses = []
list_next_tree_glasses = []
list_next_four_glasses = []
list_next_five_glasses = []

def two_glasses(next=0, min_position=1, level=0):
    if min_position==1:
        next += 1
        list_next_two_glasses.append(next)
    else:
        for a in range(0, min_position):
            next = next + min_position
            if next >= level:
                next = level
                list_next_two_glasses.append(next)
                return next
            else:
                list_next_two_glasses.append(next)
            min_position -= 1
    next += 1
    return next

def tree_glasses(next, min_position, level):
    if min_position==1:
        next += 1
        list_next_tree_glasses.append(next)
    elif min_position>1:
        min_position -= 1
        next = two_glasses(next, min_position, level)
        list_next_tree_glasses.append(next)
        if next<=level:
            while min_position > 0:
                min_position = min_position - 1
                next = two_glasses(next, min_position, level)
                if next >= level:
                    next = level
                    list_next_tree_glasses.append(next)
                    return next
                list_next_tree_glasses.append(next)
    next += 1
    return next

def four_glasses(next, min_position, level):
    if min_position==1:
        next += 1
        list_next_four_glasses.append(next)
    elif min_position>1:
        min_position -= 1
        next = tree_glasses(next, min_position, level)
        list_next_four_glasses.append(next)
        if next<=level:
            while min_position > 0:
                min_position = min_position - 1
                next = tree_glasses(next, min_position, level)
                if next >= level:
                    next = level
                    list_next_four_glasses.append(next)
                    return next
                list_next_four_glasses.append(next)
    next += 1
    return next

def five_glasses(next, min_position, level):
    if min_position==1:
        next += 1
        list_next_five_glasses.append(next)
    elif min_position>1:
        min_position -= 1
        next = four_glasses(next, min_position, level)
        list_next_five_glasses.append(next)
        if next<=level:
            while min_position > 0:
                min_position = min_position - 1
                next = four_glasses(next, min_position, level)
                if next >= level:
                    next = level
                    list_next_five_glasses.append(next)
                    return next
                list_next_five_glasses.append(next)
    next += 1
    return next

def number_of_attempts(glasses, level):
    level_next = 0
    first_level = 0
    for a in range(1, level):
        if glasses==2:
            level_next = two_glasses(first_level, a, level)
        elif glasses==3:
            level_next=tree_glasses(first_level, a, level)
        elif glasses == 4:
            level_next = four_glasses(first_level, a, level)
        elif glasses==5:
            level_next=five_glasses(first_level, a, level)
        else:
            print("Введіть від 1 до 5 стаканів")

        if level_next >= level:
            print("Максимальна кількість необхідних спроб - {}".format(a))
            print("Список можливих кидків для другого стакана:\n{}".format(list_next_two_glasses))
            print("Список можливих кидків для третього стакана:\n{}".format(list_next_tree_glasses))
            print("Список можливих кидків для четвертого стакана:\n{}".format(list_next_four_glasses))
            print("Список можливих кидків для п'ятого стакана:\n{}".format(list_next_five_glasses))
            print("____________"*5 + "\n")
            break
        else:
            list_next_two_glasses.clear()
            list_next_tree_glasses.clear()
            list_next_four_glasses.clear()
            list_next_five_glasses.clear()


for a, b in [(2, 100), (3, 100), (5, 1000)]:
    print("Для {} стаканів та {} поверхів".format(a, b))
    number_of_attempts(a, b)
```
