```python
import math

# Кількість спроб якщо використовуємо два стакани
def number_of_attempts_for_two_glasses(level = 100):
    attempt_counter = 0
    flag = True

    while flag:
        if level > attempt_counter:
            level = level - attempt_counter
            attempt_counter = attempt_counter + 1
        else:
            flag = False
    print(attempt_counter)
    for a in range(1, attempt_counter):
        print(a)
    return attempt_counter

# Кількість спроб якщо використовуємо більше ніж два стакани
def number_of_attempts_for_more_glasses(level = 100, glasses=2):
    counter_two = 0
    glasses_two = glasses
    flag = True

    while flag:
        if glasses_two > 2:
            level = math.ceil(level/2)
            print(level)
            counter_two += 1
            glasses_two -= 1
        else:
            counter_two += number_of_attempts_for_two_glasses(level)
            flag = False
    return counter_two


# Функція для використання, залежно від кількості стаканів визначаємо яку відповідь надати
def number_of_attempts(level = 100, glasses=2):
    if glasses >2:
        return number_of_attempts_for_more_glasses(level, glasses)
    elif glasses == 2:
        return number_of_attempts_for_two_glasses(level)
    elif glasses == 1:
        print("Це найгірший випадок, ми переберем лінійно усі поверхи з 1 по останній")
        return level
    else:
        print("Відсутні стакани для здійснення спроб")
        return 0

# Вивід
for a,b in [(10, 2), (100, 2), (100,3), (100, 0), (100, 5), (10, 1)]:
    print("Для хмарочоса з {} поверхів і {} склянок найгірший прохід буде по наступних поверхах:".format(a, b))
    print("Для хмарочоса з {} поверхів і {} склянок, потрібно - {} спроб".format(a, b, number_of_attempts(a, b)))
print("{} - спроб".format(number_of_attempts()), "{}".format("\nПо дефолту 100 поверхів і 2 стакани"))
```
